**Expectations**

Based on prior analysis we found that the highest POS tag for NNP (Proper Noun, Singular) with a count of 3774. This indicates a significant presence of proper nouns in the text, which are likely to be names of specific people, places, or entities. Given this context, we can set a score expectation for **polarity** and **subjectivity** in the sentiment analysis as follows:

**Polarity Score Expectation:** Since NNP tags often represent specific names of entities, events, or locations, we can expect the polarity score to be close to zero or slightly positive. Proper nouns, being neutral in nature, are less likely to express strong positive or negative sentiment. Therefore, we can anticipate a polarity score ranging from **-0.2 to +0.2**, indicating a mostly neutral sentiment in the text.

Subjectivity Score Expectation: Proper nouns typically convey factual information rather than personal opinions. Therefore, we can expect the subjectivity score to be relatively low, closer to 0. This suggests that the text is more focused on presenting factual information rather than expressing personal viewpoints. As such, the subjectivity score can be expected to fall within the range of **0 to 0.3**

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
# Open the file
file = open('key_events_20th_century.txt', 'r', encoding="utf-8")

# Read the contents
text = file.read()
```

```python
# Tokenize the text
words = nltk.word_tokenize(text)

# Remove punctuation
words = [word.lower() for word in words if word.isalpha()]

# Remove stop words
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]

# Join the filtered words back into a text
filtered_text = " ".join(filtered_words)

# Perform sentiment analysis on the filtered text
text_sentiment = TextBlob(filtered_text)
polarity_score = text_sentiment.sentiment.polarity
subjectivity_score = text_sentiment.sentiment.subjectivity

# Print the sentiment scores
print("Polarity Score:", polarity_score)
print("Subjectivity Score:", subjectivity_score)


# Create a bar plot for sentiment scores
plt.figure(figsize=(8, 5))
bars=plt.bar(["Polarity", "Subjectivity"], [polarity_score, subjectivity_score], color=['blue', 'orange'])
plt.xlabel("Sentiment Aspect")
plt.ylabel("Sentiment Score")
plt.title("Sentiment Analysis of Sample Text")

# Add data labels to the bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, "{:.2f}".format(height), ha='center', va='bottom')

plt.show()
plt.show()
```

**Inferences**


The provided analysis of the sentiment scores is 0.049. The polarity score being positive but very close to 0 indicates that the text is slightly positive or neutral in sentiment. This means that the overall sentiment of the text is not strongly positive or negative, but rather leans slightly towards positivity or neutrality.

On the other hand, the subjectivity score being 0.35 suggests that the text contains some level of personal opinion but is still relatively close to being objective. This means that while there is some subjectivity or personal opinion expressed in the text, it is not overly opinionated and still maintains a degree of objectivity.
